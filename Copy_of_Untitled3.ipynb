## MULTIVARIATE ANALYSIS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
# data1 = ...

# Selecting numeric columns
numeric_cols = data1.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr = numeric_cols.corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Numeric Variables')
plt.show()

# Visualize histograms for numeric columns
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(data1[col], bins=20, kde=True)
    plt.title(f'{col.capitalize()} Histogram')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Selecting categorical columns
categorical_cols = data1.select_dtypes(include=['object'])

# Get the top 10 most frequent values for each categorical column
top_values = {}
for col in categorical_cols:
    top_values[col] = data1[col].value_counts().nlargest(10).index.tolist()

# Visualize boxplots for each categorical variable vs. 'price' (only top 10)
for col in categorical_cols:
    if col in top_values:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=col, y='price', data=data1[data1[col].isin(top_values[col])])
        plt.title(f'{col.capitalize()} vs. Price (Top 10 Values)')
        plt.xticks(rotation=45)
        plt.show()
```
